```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year= [1950, 1970, 1990, 2010]
pop = [2.5, 3.7, 5.3, 7.0]
```

```python
plt.plot(year, pop)
plt.show()
```

```python
#scatter plot
#shows correlation 
```

```python
plt.scatter(year, pop)
plt.xscale('log') #puts the x-axis on a logarithmic scale 
plt.show()
```

```python
#histograms
#used for distrubitng datas in bins aka helps in visualizing the distribution of data
```

```python
life_exp = [76.423, 72.301, 42.731, 43.828]
```

```python
plt.hist(life_exp, bins = 5) # default number of bins is 10
plt.show()
```

```python
#customization
```

```python
plt.scatter(x=year, y=pop, s=np.array(life_exp)*5, c = ['red', 'yellow','green', 'black'], alpha = 0.8 )
year_title = 'Year'
pop_title = 'Population'

plt.xlabel(year_title)
plt.ylabel(pop_title)

plt.title('population vs year')

tick_val = [1,2,3,4,5,6,7]
tick_lab = ['1year','2year','3year', '4year' ,'5year', '6year', '7year']

plt.yticks(tick_val, tick_lab)
plt.show()
```
